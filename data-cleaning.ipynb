## Type enforcement

Convert IDs to strings, timestamps to datetimes, numeric metrics to numeric types, and categorical objects to categorical types to make grouping and plotting reliable.

Print what categories exist for each categorical column.

```python
# orders.csv
ORDER_ID_COLS = {"transaction_id", "session_key"}
ORDER_DATE_COLS = {"purchase_ts"}
ORDER_FLOAT_COLS = {"revenue"}
ORDER_INT_COLS = {"units", "item_lines", "distinct_items"}
ORDER_CATEGORICAL_COLS = {"currency"}

# Cast in place
for c in df_orders.columns:
    if c in ORDER_ID_COLS:
        df_orders[c] = df_orders[c].astype("string")
    elif c in ORDER_DATE_COLS:
        df_orders[c] = pd.to_datetime(df_orders[c], errors="coerce")
    elif c in ORDER_FLOAT_COLS:
        df_orders[c] = pd.to_numeric(df_orders[c], errors="coerce").astype("Float64")
    elif c in ORDER_INT_COLS:
        df_orders[c] = pd.to_numeric(df_orders[c], errors="coerce").astype("Int64")
    elif c in ORDER_CATEGORICAL_COLS:
        df_orders[c] = df_orders[c].astype("category")

print("categories of the `currency` column:", list(df_orders["currency"].cat.categories))
```

```python
# purchases.csv
ID_COLS = {"user_pseudo_id", "session_key", "transaction_id", "item_id"}
DATE_COLS = {"ts"}
FLOAT_COLS = {"order_value", "price", "item_revenue"}
INT_COLS = {"quantity"}
CATEGORICAL_COLS = {"currency"}
STRING_COLS = {"item_name"}

for c in df_products.columns:
    if c in ID_COLS:
        df_products[c] = df_products[c].astype("string")
    elif c in DATE_COLS:
        df_products[c] = pd.to_datetime(df_products[c], errors="coerce")
    elif c in FLOAT_COLS:
        df_products[c] = pd.to_numeric(df_products[c], errors="coerce").astype("Float64")
    elif c in INT_COLS:
        df_products[c] = pd.to_numeric(df_products[c], errors="coerce").astype("Int64")
    elif c in STRING_COLS:
        df_products[c] = df_products[c].astype("string")
    elif c in CATEGORICAL_COLS:
        df_products[c] = df_products[c].astype("category")

print("categories of the `currency` column:", list(df_products["currency"].cat.categories))
```

```python
# sessions.csv
ID_COLS = {"session_key", "user_pseudo_id"}
DATE_COLS = {"session_start_ts"}
STRING_COLS = {"landing_page_path"}
CATEGORICAL_COLS = {"session_viewport_bucket", "session_viewport_profile", "session_viewport_group", "country", "region", "device_category", "os", "browser"}

for c in df_sessions.columns:
    if c in ID_COLS:
        df_sessions[c] = df_sessions[c].astype("string")
    elif c in DATE_COLS:
        df_sessions[c] = pd.to_datetime(df_sessions[c], errors="coerce", utc=True)
    elif c in STRING_COLS:
        df_sessions[c] = df_sessions[c].astype("string")
    elif c in CATEGORICAL_COLS:
        df_sessions[c] = df_sessions[c].astype("category")


for c in (CATEGORICAL_COLS):
    if str(df_sessions[c].dtype) == "category":
        cats = list(df_sessions[c].cat.categories)
        print(f"categories of `{c}`: {cats}")
```

## Handling Missing Values

Identify columns with missing values and decide whether to drop, impute, or leave as missing to improve data quality.

`missing_report` is a helper function to identify columns with missing values and calculate what percentage of the column contains missing values.

```python
def missing_report(df, name):

    # for each column, find amount of missing values
    missing_val_count_per_col = df.isna().sum()
    
    row_count = df.shape[0] # row count

    missing_percentage = (missing_val_count_per_col/row_count) * 100 # percentage the missing counts

    missing_percentage = missing_percentage[missing_percentage > 0] # filter out complete rows

    missing_col_count = len(missing_percentage) # count how many cols have missing values

    if missing_col_count== 0:
        print(f"`{name}.csv` has 0 columns with missing values.\n")
        return

    missing_percentage = missing_percentage.sort_values(ascending=False) # sort by highest missing percentage

    print(f"`{name}.csv` has {missing_col_count} columns with missing values:")

    missing_df = missing_percentage.reset_index()
    missing_df.columns = ["column", "missing_%"]
    display(missing_df)


missing_report(df_orders, "orders")
missing_report(df_products, "purchases")
missing_report(df_sessions, "sessions")
missing_report(df_checkout, "checkout_dropoff")
missing_report(df_funnel, "events_funnel")
```

## Check for negatives

Ensure there are no negative values in columns that shouldn't have negative values.

```python
def negative_report(df, name):
    """
    If there are numeric columns, check for negatives.
    If a df has no numeric columns, skip.
    Note: The current datasets shouldn't have any values that should be negative, so we can check all columns.
    """
    num_cols = df.select_dtypes(include=["number"]).columns
    if len(num_cols) == 0:
        return
    print(f"\n{name}")
    for c in num_cols:
        neg_count = int((df[c] < 0).sum())
        print(f"negatives in column `{c}`:", neg_count)

negative_report(df_orders, "orders.csv")
negative_report(df_products, "purchases.csv")
negative_report(df_sessions, "sessions.csv")
negative_report(df_checkout, "checkout_dropoff.csv")
negative_report(df_funnel, "events_funnel.csv")
```

## Merge rows and Join coverage checks

Before EDA, verify that the merges used in `eda_part1.py` are safe.

All session keys in `df_sessions.session_key` were established to be unique in the 'Check for Duplicate Keys' section.

Therefore, we should check what percent of `df_orders` and `df_checkout`rows with a session_key exist in `df_sessions`. 100% is ideal.

```python


# orders --> sessions coverage
orders_with_key = df_orders["session_key"].notna().sum()  # count of all rows in df_orders where session key is not null

orders_session_keys = df_orders.loc[df_orders["session_key"].notna(), "session_key"]  # all non-null session_key values in orders

orders_keys_exist_in_sessions = orders_session_keys.isin(df_sessions["session_key"])  # keep the rows where the session key exists in df_sessions

orders_linked = orders_keys_exist_in_sessions.sum()  # count of all rows that can be merged

if orders_with_key == orders_linked:
    print(f"every session key in `df_orders` can be found in `df_sessions`. 100% coverage")
else:
    merge_percentage = (orders_linked / orders_with_key) * 100
    print(f"not every session key in `df_orders` can be found in `df_sessions`. {merge_percentage:.2f}% coverage.")



# checkout --> sessions coverage
checkout_with_key = df_checkout["session_key"].notna().sum()  # count of all rows in df_checkout where session key is not null

checkout_session_keys = df_checkout.loc[df_checkout["session_key"].notna(), "session_key"]  # all non-null session_key values in checkout

checkout_keys_exist_in_sessions = checkout_session_keys.isin(df_sessions["session_key"])  # keep the rows where the session key exists in df_sessions

checkout_linked = checkout_keys_exist_in_sessions.sum()  # count of all rows that can be merged

if checkout_with_key == checkout_linked:
    print(f"every session key in `df_checkout` can be found in `df_sessions`. 100% coverage")
else:
    merge_percentage = (checkout_linked / checkout_with_key) * 100 if checkout_with_key else 0.0
    print(f"not every session key in `df_checkout` can be found in `df_sessions`. {merge_percentage:.2f}% coverage.")
```
